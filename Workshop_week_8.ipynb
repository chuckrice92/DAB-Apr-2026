# 2. Data Cleaning

## Handling Missing Values

```python
customers.isnull
```

```python
customers['registration_date'] = customers['registration_date'].bfill()
customers['registration_date']
```

```python
inventory.isnull
```

```python
products.isnull
```

```python
sales.isnull
```

```python
stores.isnull
```

```python
stores['opening_date'].bfill()
```

### Replacing numeric values with mean

```python
sales['store_id'].fillna(sales['store_id'].mean())
```

```python
products['weight'].fillna(products['weight'].mean())
```

```python
stores['size_sqft'].fillna(stores['size_sqft'].mean())
```

### Replacing categorical columns

```python
customers['phone'] = customers['phone'].fillna(customers['phone'].mode()[0])
customers['phone']
```

```python
sales['payment_method'] = sales['payment_method'].fillna(sales['payment_method'].mode()[0])
sales['payment_method']
```

```python
customers['zip_code'] = customers['zip_code'].fillna(customers['zip_code'].mode()[0])
customers['zip_code']
```

### Date columns with missing values

```python
customers['registration_date'] = customers['registration_date'].bfill()
customers['registration_date']
```

```python
inventory['last_restock_date'].ffill()
```

```python
stores['opening_date'].bfill()
```

```python
sales['date'] = sales['date'].ffill()
sales['date']
```
