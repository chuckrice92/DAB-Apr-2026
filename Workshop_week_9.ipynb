```python
import numpy as np
import pandas as pd
```

# 1. DateTime Operations

## 1. Converting and Working with Dates

```python
customers = pd.read_csv('customers_wk9_thurs.csv')
customers
```

```python
customers['registration_date'] = pd.to_datetime(customers['registration_date'])
customers['registration_date']
```

```python
products = pd.read_csv('products_wk9_thurs.csv')
products
```

```python
products['launch_date'] = pd.to_datetime(products['launch_date'])
products['launch_date']
```

```python
sales = pd.read_csv('sales_wk9_thurs.csv')
sales
```

```python
sales['transaction_date'] = pd.to_datetime(sales['transaction_date'])
sales['transaction_date']
```

```python
sales['year'] = sales['transaction_date'].dt.year
sales['month'] = sales['transaction_date'].dt.month_name()
sales['day_of_week'] = sales['transaction_date'].dt.day_name()
sales['year']
```

```python
sales['month']
```

```python
sales['day_of_week']
```

```python
earliest = sales['transaction_date'].min()
latest = sales['transaction_date'].max()

print("Earliest date:", earliest)
print("Latest date:", latest)
```

## 2. Basic Time Analysis

```python
daily_counts = sales['day_of_week'].value_counts()
print(daily_counts)
print('busiest_day_of_week:', daily_counts.idxmax())
```

```python
Monthly_counts = sales['month'].value_counts()
print(Monthly_counts)
Monthly_counts.idxmax()
```

```python
yearly_counts = sales['year'].value_counts()
print(yearly_counts)
```

# 2. Groupby and Aggregation

## 1. Product Analysis

```python
average_price = products.groupby('category')['price'].mean()
print(average_price)
```

```python
product_count = products.groupby('category')['product_name'].value_counts()
print(product_count)
```

## 2. Customer Analysis

```python
average_age = customers.groupby('state')['age'].mean()
print(average_age)
```

```python
customers_total = sales.groupby('customer_id')['total_amount'].sum()
print(customers_total)
```

```python
purchase_count = sales['customer_id'].value_counts()
print(purchase_count)
```

```python
top_customers = purchase_count.head()
print(top_customers)
```

## 3. Sales Analysis

```python
total_sales_by_month = sales.groupby('month')['total_amount'].sum()
print(total_sales_by_month)
```

```python
avg_daily_sales = sales.groupby('day_of_week')['total_amount'].mean()
print(avg_daily_sales)
```

```python
revenue = products.groupby('category')['price'].sum().sort_values()
print(revenue)
```

# 3. Basic Data Insights

```python
counts = customers['state'].value_counts()
print(counts)
top_state = counts.max()
print(top_state)
```

```python
top_5_customers = sales.groupby('customer_id')['total_amount'].sum().sort_values(ascending=False).head(5)
print(top_5_customers)
```

```python
daily_average = sales.groupby('day_of_week')['total_amount'].mean().sort_values(ascending=False).head(1)
print(daily_average)
```
